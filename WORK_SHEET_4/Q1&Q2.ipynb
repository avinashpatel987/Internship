```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
```

```python
s=Service("C:\chromedriver_win32\chromedriver.exe")
driver=webdriver.Chrome(service=s)
driver.get('https://www.amazon.in/')
```

```python
product=input("Enter the Praduct Name you Want To search:")
```

```python
driver.find_element(By.ID,"twotabsearchtextbox").send_keys(product)
```

```python
btn1=driver.find_element(By.XPATH,"//*[@id='nav-search-submit-button']")
btn1.click()
```

```python
avl=[]
delivery=[]
star=[]
url=[]
name=[]
nor=[]
def text(val,which):
    if len(val) == 0:
        which.append("-")
    else:
        for i in val:
            temp=i.text
            which.append(temp)
loop=driver.find_elements(By.XPATH,"//*[@class='a-size-medium a-color-base a-text-normal']")
for i in range(1):
    prod=2
    for i in range(len(loop)):
        xp1='/html/body/div[1]/div[2]/div[1]/div[1]/div/span[3]/div[2]/div['+str(prod)+']/div/span/div/div/div/div/div[2]/div[1]/div/div/span/a/div/img'
        xp2='/html/body/div[1]/div[2]/div[1]/div[1]/div/span[3]/div[2]/div['+str(prod)+']/div/span/div/div/div[2]/div[1]/div/div/span/a/div/img'
        xp3='/html/body/div[1]/div[2]/div[1]/div[1]/div/span[3]/div[2]/div['+str(prod)+']/div/span/div/div/div/div/span/a/div/img'
        xp4='/html/body/div[1]/div[2]/div[1]/div[1]/div/span[3]/div[2]/div['+str(prod)+']/div/span/div/div/span/a/div/img'
        try:
            namclick=driver.find_element(By.XPATH,xp1)
            namclick.click() 
            time.sleep(2)
            driver.switch_to.window(driver.window_handles[1])
            a=driver.find_elements(By.XPATH,"//*[@id='availability']/span")
            text(a,avl)
            d=driver.find_elements(By.XPATH,"//*[@id='PAY_ON_DELIVERY']/div[2]/a")
            text(d,delivery)
            s=driver.find_elements(By.XPATH,"//*[@class='a-size-medium a-color-base']")
            text(s,star)
            r=driver.find_elements(By.XPATH,"/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[5]/div/span[3]/a")
            text(r,nor)
            n=driver.find_elements(By.XPATH,"//*[@id='productTitle']")
            text(n,name)
            url.append(driver.current_url)
            driver.close()
            driver.switch_to.window(driver.window_handles[0])
        except:
            try:
                namclick=driver.find_element(By.XPATH,xp2)
                namclick.click() 
                time.sleep(2)
                driver.switch_to.window(driver.window_handles[1])
                a=driver.find_elements(By.XPATH,"//*[@id='availability']/span")
                text(a,avl)
                d=driver.find_elements(By.XPATH,"//*[@id='PAY_ON_DELIVERY']/div[2]/a")
                text(d,delivery)
                s=driver.find_elements(By.XPATH,"//*[@class='a-size-medium a-color-base']")
                text(s,star)
                r=driver.find_elements(By.XPATH,"/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[5]/div/span[3]/a")
                text(r,nor)
                n=driver.find_elements(By.XPATH,"//*[@id='productTitle']")
                text(n,name)
                url.append(driver.current_url)
                driver.close()
                driver.switch_to.window(driver.window_handles[0])
            except:
                try:
                    namclick=driver.find_element(By.XPATH,xp3)
                    namclick.click() 
                    time.sleep(2)
                    driver.switch_to.window(driver.window_handles[1])
                    a=driver.find_elements(By.XPATH,"//*[@id='availability']/span")
                    text(a,avl)
                    d=driver.find_elements(By.XPATH,"//*[@id='PAY_ON_DELIVERY']/div[2]/a")
                    text(d,delivery)
                    s=driver.find_elements(By.XPATH,"//*[@class='a-size-medium a-color-base']")
                    text(s,star)
                    r=driver.find_elements(By.XPATH,"/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[5]/div/span[3]/a")
                    text(r,nor)
                    n=driver.find_elements(By.XPATH,"//*[@id='productTitle']")
                    text(n,name)
                    url.append(driver.current_url)
                    driver.close()
                    driver.switch_to.window(driver.window_handles[0])
                except:
                    try:
                        namclick=driver.find_element(By.XPATH,xp4)
                        namclick.click() 
                        time.sleep(2)
                        driver.switch_to.window(driver.window_handles[1])
                        a=driver.find_elements(By.XPATH,"//*[@id='availability']/span")
                        text(a,avl)
                        d=driver.find_elements(By.XPATH,"//*[@id='PAY_ON_DELIVERY']/div[2]/a")
                        text(d,delivery)
                        s=driver.find_elements(By.XPATH,"//*[@class='a-size-medium a-color-base']")
                        text(s,star)
                        r=driver.find_elements(By.XPATH,"/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[5]/div/span[3]/a")
                        text(r,nor)
                        n=driver.find_elements(By.XPATH,"//*[@id='productTitle']")
                        text(n,name)
                        url.append(driver.current_url)
                        driver.close()
                        driver.switch_to.window(driver.window_handles[0])
                    except:
                        url.append("-")
                        name.append("-")
                        nor.append("-")
                        delivery.append("-")
                        star.append("-")
                        avl.append("-")
        prod+=1
    time.sleep(2)
    next=driver.find_element(By.XPATH,"//li[@class='a-last']")
    next.click()
    time.sleep(2)
       
```
