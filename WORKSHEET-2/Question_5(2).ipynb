# Importing Library

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import string
import re
```

# Getting URL of the Website

```python
url='https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting'
cont=requests.get(url)
htmlContent=cont.content
```

# Creating The soup

```python
soup=BeautifulSoup(htmlContent,'html.parser')
print(soup.prettify)
```

# Names

```python
head=soup.find_all('td',class_="table-body__cell rankings-table__name name")
head
```

```python
D=[]
for i in head:
    i=i.get_text().replace("\n","")
    i=i.strip(" ")
    D.append(i)
D
```

```python
head2=soup.find_all('div',class_="rankings-block__banner--name-large")
head2
```

```python
D2=[]
for i in head2:
    i=i.get_text().replace("\n","")
    i=i.strip(" ")
    D2.append(i)
D2
```

```python
wpn=[]
wpn=D2+D
wpn
```

```python
wpnact=wpn[:10]
```

```python
cont=[]
cont=D[0:10]
```

# Country

```python
bn2=soup.find_all('div',class_='rankings-block__banner--nationality')
bn2
```

```python
cn2=[]
for i in bn2:
    i=i.get_text().replace("\n","")
    i=i.strip(" ")
    cn2.append(i)
cn2
```

```python
cnr=[]
cnr=cn2+cn
cnr
```

```python
cnact=cnr[:10]
```
