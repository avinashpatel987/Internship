# Write a python program to scrape monument name, monument description, image url about top 10 monuments from 'https://www.puredestinations.co.uk/top-10-famous-monuments-to-visit-in-india/'

# Importing Library

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import string
import re
```

```python
url='https://www.puredestinations.co.uk/top-10-famous-monuments-to-visit-in-india/'
c=requests.get(url)
htmlContent=c.content
```

```python
print(htmlContent)
```

# Soup

```python
soup=BeautifulSoup(htmlContent,'html.parser')
print(soup.prettify)
```

# Names

```python
head=soup.find_all('strong')
head
```

```python
names=[]
for i in head:
    i=i.get_text().replace("\n","")
    i=i.strip(" ")
    names.append(i)
names
```

```python
mn=[]
mn=names[:10]
```

# Description

```python
head2=soup.find_all('p')
head2
```

```python
d=[]
for i in head2:
    i=i.get_text().replace("\n","")
    i=i.strip(" ")
    d.append(i)
d
```

```python
desc=[]
desc=d[5::3]
```

# URL

```python
links=soup.select('img[src^="http://www.puredestinations.co.uk/wp-content/uploads/2016/11"]')
print(links)
```

```python
images=[]
for i in range(len(links)):
    images.append(links[i]['src'])
images
```
