# Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame.

# Importing Library

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import string
import re
```

# Creating the Soup

```python
url='https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc'
c=requests.get(url)
htmlContent=c.content
```

```python
print(htmlContent)
```

# Parsing the soup

```python
soup=BeautifulSoup(htmlContent,'html.parser')
print(soup.prettify)
```

# Finding all the names of the Movies

```python
head=soup.find_all('h3',class_='lister-item-header')
head
```

```python
movieNames=[]
for i in head:
    i=i.get_text().replace("\n","")
    i=i.strip(" ")
    movieNames.append(i)
movieNames
```

# Removing all the Extras From Names from the Movies

```python
pattern = r'[0-9,.,()]'
Names=[]
for i in range(len(movieNames)):
    Names.append(re.sub(pattern, '', movieNames[i]))
Names
```

# Getting All the Year Of Release

```python
pattern = r'[a-z,A-Z]'
Year=[]
for i in range(len(movieNames)):
    Year.append(re.sub(pattern, '', movieNames[i]))
Year
```

# Finding all the Rating of the movies

```python
r=soup.find_all("strong")
ratings=[]
for i in r:
    ratings.append(i.get_text().replace("\n",""))
ratings    
```

```python
del ratings[0:2]
```

# Creating the Dataset

```python
dataset=pd.DataFrame()
dataset['Movie Names/YOR']=Names
dataset['Ratings']=ratings
dataset['Year']=Year
dataset.head()
```
