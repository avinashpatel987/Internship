```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import urllib
import os
from selenium.webdriver.common.action_chains import ActionChains
```

```python
s=Service("C:\chromedriver_win32\chromedriver.exe")
driver=webdriver.Chrome(service=s)
driver.get('https://archive.ics.uci.edu/')
```

```python
btn1=driver.find_element(By.XPATH,'/html/body/table[2]/tbody/tr/td/span/b/a')
btn1.click()
```

```python
name=[]
n=driver.find_elements(By.XPATH,"//p[@class='normal']/b/a")
text(n,name)
```

```python
datatypes=[]
defaulttasks=[]
attributesTypes=[]
instance=[]
attribute=[]
year=[]

def text(val,which):
        for i in val:
            temp=i.text
            if temp.strip(' ')=='':
                which.append("-")
            else:
                which.append(temp)
                len(temp)

for i in range(2,610):
    for j in range(2,8):
        if j==2:
            dt=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td['+str(j)+']/p')
            text(dt,datatypes)
        elif j==3:
            deftsk=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td['+str(j)+']/p')
            text(deftsk,defaulttasks)
        elif j==4:
            attps=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td['+str(j)+']/p')
            text(attps,attributesTypes)
        elif j==5:
            inst=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td['+str(j)+']/p')
            text(inst,instance)
        elif j==6:
            att=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td['+str(j)+']/p')
            text(att,attribute)
        else:
            y=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td['+str(j)+']/p')
            text(y,year)
```

```python
dataset=pd.DataFrame()
dataset['Name']=name[:608]
dataset['Datatypes']=datatypes
dataset['Default Task']=defaulttasks
dataset['Attribute Types']=attributesTypes
dataset['Instances']=instance
dataset['Attributes']=attribute
dataset['Year']=year
dataset
```
