```python
name=[]
ram=[]
display=[]
camera=[]
battery=[]
processor=[]
url=[]
def text(val,which):
    if len(val) == 0:
        which.append("-")
    else:
        for i in val:
            temp=i.text
            which.append(temp)
loop=driver.find_elements(By.XPATH,"//*[@class='col col-7-12']")
prod=2
for i in range(len(loop)):
    xp1='//*[@id="container"]/div/div[3]/div[1]/div[2]/div['+str(prod)+']/div/div/div/a/div[1]/div[1]/div/div/img'
    xp2='//*[@id="container"]/div/div[3]/div[1]/div[2]/div['+str(prod)+']/div/div/div/a/div[1]/div[1]/div[1]/div/img'
    try:
        namclick=driver.find_element(By.XPATH,xp1)
        namclick.click() 
        time.sleep(2)
        driver.switch_to.window(driver.window_handles[1])
        n=driver.find_elements(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[2]/div/div[1]/h1/span')
        text(n,name)
        r=driver.find_elements(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[7]/div[1]/div/div[2]/ul/li[1]')
        text(r,ram)
        d=driver.find_elements(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[7]/div[1]/div/div[2]/ul/li[2]')
        text(d,display)
        c=driver.find_elements(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[7]/div[1]/div/div[2]/ul/li[3]')
        text(c,camera)
        b=driver.find_elements(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[7]/div[1]/div/div[2]/ul/li[4]')
        text(b,battery)
        p=driver.find_elements(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[7]/div[1]/div/div[2]/ul/li[5]')
        text(p,processor)
        url.append(driver.current_url)
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
    except:
        try:
            namclick=driver.find_element(By.XPATH,xp2)
            namclick.click() 
            time.sleep(2)
            driver.switch_to.window(driver.window_handles[1])
            n=driver.find_elements(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[2]/div/div[1]/h1/span')
            text(n,name)
            r=driver.find_elements(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[7]/div[1]/div/div[2]/ul/li[1]')
            text(r,ram)
            d=driver.find_elements(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[7]/div[1]/div/div[2]/ul/li[2]')
            text(d,display)
            c=driver.find_elements(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[7]/div[1]/div/div[2]/ul/li[3]')
            text(c,camera)
            b=driver.find_elements(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[7]/div[1]/div/div[2]/ul/li[4]')
            text(b,battery)
            p=driver.find_elements(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[7]/div[1]/div/div[2]/ul/li[5]')
            text(p,processor)
            url.append(driver.current_url)
            driver.close()
            driver.switch_to.window(driver.window_handles[0])
        except:
            url.append("-")
            name.append("-")
            nor.append("-")
            delivery.append("-")
            star.append("-")
            avl.append("-")
    prod+=1
```

```python
dataset=pd.DataFrame()
dataset['Product Name']=name
dataset['Ram']=ram
dataset['Camera']=camera
dataset['Display']=display
dataset['Processor']=processor
dataset['Battery']=battery
dataset['URL']=url
dataset.head()
```
