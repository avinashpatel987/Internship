```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import urllib
import os
```

```python
s=Service("C:\chromedriver_win32\chromedriver.exe")
driver=webdriver.Chrome(service=s)
driver.get('https://www.digit.in/')
```

```python
btn1=driver.find_element(By.XPATH,"/html/body/div[1]/div/div[3]/div[2]/a/img")
btn1.click()
```

```python
product=input("Enter what to search:")
```

```python
driver.find_element(By.XPATH,'//*[@id="globalPageSearchText"]').send_keys(product)
```

```python
from selenium.webdriver.common.keys import Keys
driver.find_element(By.XPATH,'//*[@id="globalPageSearchText"]').send_keys(Keys.ENTER)
```

```python
name=[]
display=[]
memory=[]
processor=[]
storage=[]
power=[]
url=[]
def text(val,which):
    if len(val) == 0:
        which.append("-")
    else:
        for i in val:
            temp=i.text
            which.append(temp)
loop=driver.find_elements(By.XPATH,"//*[@class='searchProduct-desc']")
prod=3
for i in range(len(loop)):
    xp1='/html/body/div[6]/div[2]/div[2]/div['+str(prod)+']/a/div'
    try:
        namclick=driver.find_element(By.XPATH,xp1)
        namclick.click() 
        time.sleep(1)
        btn=driver.find_element(By.XPATH,"/html/body/div[3]/div[2]/div[4]/ul/li[2]/a")
        btn.click()
        time.sleep(1)
        n=driver.find_elements(By.XPATH,'/html/body/div[3]/div[2]/div[2]/h1')
        text(n,name)
        d=driver.find_elements(By.XPATH,'//*[@id="specs-display"]/div[2]')
        text(d,display)
        m=driver.find_elements(By.XPATH,'//*[@id="specs-memory"]/div[2]/table/tbody')
        text(m,memory)
        p=driver.find_elements(By.XPATH,'//*[@id="specs-processor"]/div[2]/table/tbody')
        text(p,processor)
        s=driver.find_elements(By.XPATH,'//*[@id="specs-storage"]/div[2]/table/tbody')
        text(s,storage)
        po=driver.find_elements(By.XPATH,'//*[@id="specs-power"]/div[2]/table/tbody')
        text(po,power)
        url.append(driver.current_url)
        driver.back()
        driver.back()
    except:
        url.append("-")
        name.append("-")
        memory.append("-")
        display.append("-")
        storage.append("-")
        power.append("-")
        processor.append("-")
        driver.back()
        driver.back()
    prod+=1
```
