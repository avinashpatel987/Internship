# Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

# Importing Library

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import string
import re
import urllib.request
```

# Header For the Website To Think That We Are Not A Robot But Human

```python
header= {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
```

# Getting URL

```python
url='https://www.amazon.in/s?k=mobile+phoen+under+20000&dc&ref=a9_asc_1'
c=requests.get(url,headers=header)
htmlContent=c.content
```

```python
print(htmlContent)
```

# Soup

```python
soup=BeautifulSoup(htmlContent,'html.parser')
print(soup.prettify)
```

# Products

```python
head=soup.find_all('span',class_='a-size-medium a-color-base a-text-normal')
head
```

```python
p=[]
for i in head:
    i=i.get_text().replace("\n","")
    i=i.strip(" ")
    p.append(i)
p
```

```python
products=[]
products=p[1:21]
```

# Price

```python
head2=soup.find_all('span',class_='a-price-whole')
head2
```

```python
rs=[]
for i in head2:
    i=i.get_text().replace("\n","")
    i=i.strip(" ")
    rs.append(i)
rs
```

```python
rupee=[]
rupee=rs[:20]
```

# Image URL

```python
head3=soup.find_all('img',class_='s-image')
head3
```

```python
links=soup.select('img[src^="https://m.media-amazon.com/images/I"]')
print(links)
```

```python
images=[]
for i in range(len(links)):
    images.append(links[i]['src'])
images
```

```python
urlimg=[]
urlimg=images[:20]
```

# Rating

```python
head4=soup.find_all('span',class_='a-icon-alt')
head4
```

```python
rat=[]
for i in head4:
    i=i.get_text().replace("\n","")
    i=i.strip(" ")
    rat.append(i)
rat
```

```python
stars=[]
stars=rat[:20]
```

# Database

```python
dataset=pd.DataFrame()
dataset['Product']=products
dataset['Price']=rupee
dataset['Image_URL']=urlimg
dataset['Rating']=stars
dataset.head(10)
```
