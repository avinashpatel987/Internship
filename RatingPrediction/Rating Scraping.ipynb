```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import urllib
import os
from selenium.webdriver import ActionChains

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
s=Service("C:\chromedriver_win32\chromedriver.exe")
driver=webdriver.Chrome(service=s)
driver.get('https://www.amazon.com')
```

```python
def text(val,which):
    temp=[]
    if len(val) == 0:
        which.append("-")
    else:
        for i in val:
            temp=i.text
            which.append(temp)
```

```python
def product_click(i):
    product_url1='//*[@id="search"]/div[1]/div[1]/div/span[3]/div[2]/div['+str(i)+']/div/div/div/div/div/div/div/div[2]/div/div/div[1]/h2/a/span'
    product_url2='//*[@id="search"]/div[1]/div[1]/div/span[3]/div[2]/div['+str(i)+']/div/div/div/div/div/div[2]/div/div/div[1]/h2/a/span'
    product_url3='//*[@id="search"]/div[1]/div[1]/div/span[3]/div[2]/div['+str(i)+']/div/div/div/div[2]/div[2]/div[2]/div/div[1]/h2/a/span'
    product_url4='//*[@id="search"]/div[1]/div[1]/div/span[3]/div[2]/div['+str(i)+']/div/div/div/div/div/div[2]/div/div/div[1]/h2/a/span'
    try:             
        product=WebDriverWait(driver, 2).until(EC.presence_of_element_located((By.XPATH,product_url1)))
        product.click()
    except:
        try:
            product=WebDriverWait(driver, 2).until(EC.presence_of_element_located((By.XPATH,product_url2)))
            product.click()
        except:
            try:
                product=WebDriverWait(driver, 2).until(EC.presence_of_element_located((By.XPATH,product_url3)))
                product.click()
            except:
                product=WebDriverWait(driver, 2).until(EC.presence_of_element_located((By.XPATH,product_url4)))
                product.click()
```

```python
def star_count(len_,j):
    for i in range(len_):
        if j == 1:
            st.append(5)
        elif j==2:
            st.append(4)
        elif j==3:
            st.append(3)
        elif j==4:
            st.append(2)
        else:
            st.append(1)
```

```python
def star_next_page(j):
    try:
        for n in range(5):
            clenbefore=len(c)
            comm=d=driver.find_elements(By.XPATH,'//div[@class="a-section celwidget"]/div[4]')
            num_=text(comm,c)
            time.sleep(.5)
            clenafter=len(c)
            len_=clenafter-clenbefore
            time.sleep(1)
            star_count(len_,j)
            time.sleep(1)
            nxt=WebDriverWait(driver, 1).until(EC.presence_of_element_located((By.XPATH,'//*[@id="cm_cr-pagination_bar"]/ul/li[2]/a')))
            nxt.click()
    except:
        pass
```

```python
def see_all_reviews():
    seeAllReviews=WebDriverWait(driver, 2).until(EC.presence_of_element_located((By.XPATH,'//*[@id="cr-pagination-footer-0"]/a | //*[@id="reviews-medley-footer"]/div[2]/a')))
    seeAllReviews.click()
    for j in range(1,6):
        try:
            str_url='//*[@id="histogramTable"]/tbody/tr['+str(j)+']/td[1]/span[1]/a'
            str_=WebDriverWait(driver, 2).until(EC.presence_of_element_located((By.XPATH,str_url)))
            str_.click()
            time.sleep(1)
            star_next_page(j)
        except:
            pass
```
