# 1. Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

# Importing Library

```python
from bs4 import BeautifulSoup
import requests
```

# Getting content

```python
url='https://en.wikipedia.org/wiki/Main_Page'
c=requests.get(url)
htmlContent=c.content
```

# Printing HTML Of the Page

```python
print(htmlContent)
```

# Parsing the HTML

```python
soup=BeautifulSoup(htmlContent,'html.parser')
print(soup.prettify)
```

# Getting All th HTML tags

```python
head=[]
h=['h1','h2','h3','h4','h5','h6']
for i in h:
    head.append(soup.find_all(i))
head
```
