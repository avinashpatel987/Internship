# Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame.

# Importing Library

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import string
import re
```

# Getting the Site URL

```python
url='https://www.imdb.com/india/top-rated-indian-movies/'
c=requests.get(url)
htmlContent=c.content
```

```python
print(htmlContent)
```

# Creating the Soup

```python
soup=BeautifulSoup(htmlContent,'html.parser')
print(soup.prettify)
```

# Name OF Movies

```python
head=soup.find_all('td',class_='titleColumn')
head
```

```python
movieNames=[]
for i in head:
    i=i.get_text().replace("\n","")
    i=i.strip(" ")
    movieNames.append(i)
movieNames
```

```python
pattern = r'[0-9,.,()]'
Names=[]
for i in range(len(movieNames)):
    Names.append(re.sub(pattern, '', movieNames[i]))
Names
```

# Getting All the Year Of Release

```python
pattern = r'[a-z,A-Z,/,#,!,:,_,-,.,&]'
Year=[]
for i in range(len(movieNames)):
    Year.append(re.sub(pattern, '', movieNames[i]))
Year
```

# Getting All the Rating

```python
r=soup.find_all("td",class_='ratingColumn imdbRating')
ratings=[]
for i in r:
    ratings.append(i.get_text().replace("\n",""))
ratings    
```

# Creating The Database

```python
dataset=pd.DataFrame()
dataset['Movie Names/YOR']=Names
dataset['Ratings']=ratings
dataset['Year']=Year
dataset.head()
```
