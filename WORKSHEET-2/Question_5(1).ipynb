# Importing Library

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import string
import re
```

# Getting The URL Content

```python
url='https://www.icc-cricket.com/rankings/womens/team-rankings/odi'
c=requests.get(url)
htmlContent=c.content
```

```python
print(htmlContent)
```

# Creating the soup

```python
soup=BeautifulSoup(htmlContent,'html.parser')
print(soup.prettify)
```

# Team Name

```python
head=soup.find_all('span',class_='u-hide-phablet')
head
```

```python
count=0
Team=[]
TeamNames=[]
for i in head:
    i=i.get_text()
    Team.append(i)
for i in Team:
    if(count<10):
        TeamNames.append(i)
        count+=1  
TeamNames
```

# Points

```python
r=soup.find_all('td',{"class": ["rankings-block__banner--points", "table-body__cell u-center-text"]})
r
```

```python
p=[]
for i in r:
    i=i.get_text()
    p.append(i)
p
```

```python
points=[]
points=p[0:20:2]
points
```

# Rating

```python
rating=soup.find_all('td',{"class": ["rankings-block__banner--rating u-text-right", "table-body__cell u-text-right rating"]})
rating
```

```python
rat=[]
for i in rating:
    i=i.get_text().replace("\n","")
    i=i.strip(" ")
    rat.append(i)
rat
```

```python
ratings=[]
ratings=rat[0:10]
ratings
```

# Creating The dataset

```python
dataset=pd.DataFrame()
dataset['Team Names']=TeamNames
dataset['Ratings']=ratings
dataset['Points']=points
dataset.head()
```
