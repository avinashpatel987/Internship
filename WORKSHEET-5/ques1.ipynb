```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

```python
s=Service("C:\chromedriver_win32\chromedriver.exe")
driver=webdriver.Chrome(service=s)
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
rank=[]
name=[]
uploadBy=[]
views=[]
uploadDate=[]
def text(val,which):
    if len(val) == 0:
        which.append("-")
    else:
        for i in val:
            temp=i.text
            which.append(temp)
for i in range(1,31):
    for j in range(1,6):
        if j==1:
            r=driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr['+str(i)+']/td['+str(j)+']')
            text(r,rank)
        elif j==2:
            n=driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr['+str(i)+']/td['+str(j)+']')
            text(n,name)
        elif j==3:
            ub=driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr['+str(i)+']/td['+str(j)+']')
            text(ub,uploadBy)
        elif j==4:
            v=driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr['+str(i)+']/td['+str(j)+']')
            text(v,views)
        else:
            ud=driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr['+str(i)+']/td['+str(j)+']')
            text(ud,uploadDate)
```

```python
dataset=pd.DataFrame()
dataset['Rank']=rank
dataset['Video Name']=name
dataset['Creator Name']=uploadBy
dataset['Views(Billions)']=views
dataset['Upload Date']=uploadDate
dataset.head()
```
