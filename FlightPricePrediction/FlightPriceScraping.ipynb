```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import urllib
import os
from selenium.webdriver import ActionChains

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
def text(val,which):
    if len(val) == 0:
        which.append("-")
    else:
        for i in val:
            temp=i.text
            which.append(temp)
```

# Scraping For Different Cities In Yatra.com

```python
def from_():
    from_=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'//div[@class="fsw_inputBox searchCity inactiveWidget "]')))
    from_.click()
    time.sleep(1)
    sk1=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'//*[@id="root"]/div/div[2]/div/div/div[2]/div[1]/div[1]/div[1]/div/div/div/input')))
    sk1.send_keys(i)
    time.sleep(1)
    optn=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'//*[@id="react-autowhatever-1-section-0-item-0"]/div/div[1]')))
    optn.click()
```

```python
def to_():
    to_=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'/html/body/div/div/div[2]/div/div/div[2]/div[1]/div[2]')))
    to_.click()
    time.sleep(1)
    sk2=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'/html/body/div/div/div[2]/div/div/div[2]/div[1]/div[2]/div[1]/div/div/div[1]/input')))
    sk2.send_keys(j)
    time.sleep(1)
    optn=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'//*[@id="react-autowhatever-1-section-0-item-0"]/div/div[1]')))
    optn.click()
```

```python
def search_():
    time.sleep(1)
    search=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'//*[@id="root"]/div/div[2]/div/div/div[2]/p/a')))
    search.click()
    time.sleep(3)
```

```python
state=["NEW DELHI","MUMBAI","JAMMU","CHANDIGARH","DHARAMSHALA","JAIPUR","DEHRADUN","AHMEDABAD","INDORE","BIHAR","BILASPUR","RANCHI","SIKKIM","KOLKATA","GUWAHATI","LILABARI","NAGALAND","TRIPURA","MANIPUR","MIZORAM","MAHARASTRA","TELANGANA","GOA","KARNATAKA","ANDHRA PRADESH","Thiruvananthapuram","TAMIL NADU","ODISHA"]
fromto=[]
flightName=[]
time_=[]
duration=[]
stops=[]
date=[]
price=[]
for i in state:
    for j in state:
        if i!=j:
            time.sleep(2)
            #from
            from_()
            #to
            to_()
            #blank click
            time.sleep(2)
            driver.find_element(By.XPATH,"//body").click()
            driver.find_element(By.XPATH,"//body").click()
            driver.find_element(By.XPATH,"//*[@id='root']/div/div[2]/div/div/div[2]/div[2]/div[1]/ul/li[1]").click()
            #Click Search
            time.sleep(2)
            element=driver.find_element(By.XPATH,"//*[@id='SW']/div[1]/div[1]/a/img")
            ActionChains(driver).move_to_element(element).perform()  
            search_()
            #close Adds
            try:
                adds=driver.find_element(By.XPATH,'//*[@id="root"]/div/div[2]/div[2]/div[2]/div/span')
                adds.click()
            except:
                pass
            #scrolling
            try:
                for a in range(10):
                    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
            except:
                pass
            #Info Gathering
            for z in range(2,10,2):
                time.sleep(2)
                fn=driver.find_elements(By.XPATH,'//span[@class="boldFont blackText airlineName"]')
                ti=driver.find_elements(By.XPATH,'//p[@class="blackText fontSize18 blackFont appendBottom2 makeFlex hrtlCenter"]')
                d=driver.find_elements(By.XPATH,'//div[@class="stop-info flexOne"]/p')
                s=driver.find_elements(By.XPATH,'//p[@class="flightsLayoverInfo"]')
                p=driver.find_elements(By.XPATH,'//p[@class="blackText fontSize18 blackFont white-space-no-wrap"]')
                print(len(p))
                for t in range(len(p)):
                    temp=WebDriverWait(driver, 20).until(EC.visibility_of_element_located((By.XPATH, "/html/body/div[1]/div/div[2]/header/div[1]/div/div[5]/input"))).get_attribute("title")
                    date.append(temp)
                    print(len(date))
                frto=driver.find_elements(By.XPATH,'//p[@class="darkText"]')
                text(fn,flightName)
                text(ti,time_)
                text(d,duration)
                text(s,stops)
                text(frto,fromto)
                text(p,price)
                driver.find_element(By.TAG_NAME,'body').send_keys(Keys.CONTROL + Keys.HOME)
                driver.find_element(By.TAG_NAME,'body').send_keys(Keys.CONTROL + Keys.HOME)
                try:
                    nxtdate=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'//div[@class="glider-track"]/div['+str(z)+']/a/p[1]')))
                    nxtdate.click()
                except:
                    try:
                        nx= WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'/html/body/div/div/div[2]/div[2]/div/div[2]/div[2]/div[1]/div/button[2]')))
                        nx.click()
                        nxtdate=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'//div[@class="glider-track"]/div['+str(z)+']/a/p[1]')))
                        nxtdate.click()
                    except:
                        try:
                            nx= WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'/html/body/div/div/div[2]/div[2]/div/div[2]/div[2]/div[1]/div/button[2]')))
                            nx.click()
                            nxtdate=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'//div[@class="glider-track"]/div['+str(z)+']/a/p[1]')))
                            nxtdate.click()
                        except:
                            try:
                                nx= WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'/html/body/div/div/div[2]/div[2]/div/div[2]/div[2]/div[1]/div/button[2]')))
                                nx.click()
                                nxtdate=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'//div[@class="glider-track"]/div['+str(z)+']/a/p[1]')))
                                nxtdate.click()
                            except:
                                print("ERROR!!!")
                try:
                    adds= WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'//*[@id="root"]/div/div[2]/div[2]/div[2]/div/span')))
                    adds.click()
                except:
                    pass
                try:
                    for a in range(10):
                        driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
                except:
                    pass
            #Back
            driver.close()
            s=Service("C:\chromedriver_win32\chromedriver.exe")
            driver=webdriver.Chrome(service=s)
            driver.get('https://www.makemytrip.com')
            time.sleep(3)
            driver.find_element(By.XPATH,"//body").click()
            driver.find_element(By.XPATH,"//body").click()
            
```
