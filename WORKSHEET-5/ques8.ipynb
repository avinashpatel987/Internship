```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import urllib
import os
from selenium.webdriver.common.action_chains import ActionChains
```

```python
s=Service("C:\chromedriver_win32\chromedriver.exe")
driver=webdriver.Chrome(service=s)
driver.get('https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare')
```

```python
all_=driver.find_elements(By.XPATH,"//td[@class='left']")
gen=driver.find_elements(By.XPATH,"//td[@class='last left']")
```

```python
def text(val,which):
    if len(val) == 0:
        which.append("-")
    else:
        for i in val:
            temp=i.text
            which.append(temp)
```

```python
everything=[]
genre=[] 
text(all_,everything)
text(gen,genre)
```

```python
everything
```

```python
every=[everything[i:i+5] for i in range(0,len(everything),5)]
```

```python
every[1]
```

```python
rank=[]
title=[]
author=[]
volume_seriese=[]
publisher=[]
for i in range(len(every)):
    temp=every[i]
    rank.append(temp[0])
    title.append(temp[1])
    author.append(temp[2])
    volume_seriese.append(temp[3])
    publisher.append(temp[4])
```

```python
dataset=pd.DataFrame()
dataset['Rank']=rank
dataset['Title']=title
dataset['Author']=author
dataset['Volume Seriese']=volume_seriese
dataset['Publisher']=publisher
dataset['Genre']=genre
dataset
```
